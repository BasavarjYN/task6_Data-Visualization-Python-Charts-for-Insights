```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("country_wise_latest.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
top10 = (
    df.groupby("Country/Region")["Confirmed"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
top10.plot(kind="bar")
plt.xlabel("Country")
plt.ylabel("Total Confirmed Cases")
plt.title("Top 10 Countries by Confirmed COVID-19 Cases")
plt.grid(True)
plt.show()
```

```python
region_deaths = df.groupby("WHO Region")["Deaths"].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
region_deaths.plot(kind="bar")
plt.xlabel("WHO Region")
plt.ylabel("Total Deaths")
plt.title("COVID-19 Deaths by WHO Region")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Confirmed"], bins=20)
plt.xlabel("Confirmed Cases")
plt.ylabel("Frequency")
plt.title("Distribution of Confirmed COVID-19 Cases")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["Confirmed"], df["Deaths"])
plt.xlabel("Confirmed Cases")
plt.ylabel("Deaths")
plt.title("Confirmed vs Deaths")
plt.grid(True)
plt.show()
```

```python
# INSIGHTS
# A small number of countries contribute the majority of confirmed COVID-19 cases.
# There is a strong positive relationship between confirmed cases and deaths.
# The confirmed case distribution is highly right-skewed, indicating extreme outliers.
```
